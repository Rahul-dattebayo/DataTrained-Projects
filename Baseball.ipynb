```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
lr = LinearRegression()

for i in range(0,1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train, y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    if round(mean_absolute_error(pred_train, y_train)*100,1) == round(mean_absolute_error(pred_test, y_test)*100,1):
        print('at random state', i, 'The model performs well')
        print('at random state', i)
        print('training socre is:', mean_absolute_error(pred_train, y_train)*100)
        print('testing score is:', mean_absolute_error(pred_test, y_test)*100)
```

```python
#Let's take the threshold skewness to be 0.5.

#Log Tranformation is applied to remove the skewness.

for index in x.skew().index:
    if x.skew().loc[index]>0.5:
        x[index]=np.log1p(x[index])
    if x.skew().loc[index]<-0.5:
        x[index]=np.square(x[index])
```

### Imbalanced data

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
def max_score(rc,x,y):
    max_score=0
    for i in range(35,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = i,test_size=0.20,stratify=y)
        x_train, y_train = SMOTE().fit_sample(x_train, y_train)
        rc.fit(x_train,y_train)
        pred = rc.predict(x_test)
        score = roc_auc_score(y_test,pred)
        print("auc roc score corresponding to ",r," is ",score)
        if score>max_score:
            max_score=score
            final_r = r
    print("max auc roc score corresponding to ",final_r," is ",max_score)
    return final_r
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
def max_score(rc,x,y):
    max_score=0
    for r in range(35,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r,test_size=0.20)
        rc.fit(x_train,y_train)
        pred = rc.predict(x_test)
        score = roc_auc_score(y_test,pred)
        print("auc roc score corresponding to ",r," is ",score)
        if score>max_score:
            max_score=score
            final_r = r
    print("max auc roc score corresponding to ",final_r," is ",max_score)
    return final_r
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
lg=LogisticRegression()
max_score(lg,x,y)
```

```python
from sklearn.model_selection import cross_val_score
print("Mean roc auc score for logistic classifier: ",cross_val_score(lg,x,y,cv=5,scoring="roc_auc").mean())
print("standard deviation in roc auc score for logistic classifier: ",cross_val_score(lg,x,y,cv=5,scoring="roc_auc").std())
print(cross_val_score(lg,x,y,cv=5,scoring="roc_auc"))
```

### DTC

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
max_score(dc,x,y)
```

### RFC

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
parameters={"n_estimators":[10,100,500]}
rf=RandomForestClassifier()
clf = GridSearchCV(rf, parameters, cv=5,scoring="roc_auc")
clf.fit(x,y)
clf.best_params_
```
